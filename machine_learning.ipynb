```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# array_1 = 12 * np.random.rand(100,2) + np.array([10,10])
# array_2 = 8 * np.random.rand(100,2) + np.array([8,22])
array_1 = np.random.normal((16,14), 2, (100,2))
array_2 = np.random.normal((32,22), 3, (100,2))

plt.scatter(array_1[:,0], array_1[:,1], c='b')
plt.scatter(array_2[:,0], array_2[:,1], c='r')
plt.show()
```

```python
train_data = np.concatenate([array_1, array_2])
train_data.shape
```

```python
from sklearn.mixture import GaussianMixture

gmm = GaussianMixture(n_components=2)

gmm.fit(train_data)

print(gmm.means_)
print(gmm.weights_)
print(gmm.covariances_)

plt.scatter(array_1[:,0], array_1[:,1], c='b')
plt.scatter(array_2[:,0], array_2[:,1], c='r')
plt.scatter(gmm.means_[:,0], gmm.means_[:,1], c='y')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
from itertools import product

for eps, min_samples in product([0.5,1.2,1.8,2.6,4], [4,6,10,16]):
    dbm = DBSCAN(eps=eps, min_samples=min_samples)

    dbm.fit(train_data)
    n_classes = len(list(set(dbm.labels_)))
    bruit = len([i for i in dbm.labels_ if i == -1])
    print(f'Pour eps={eps} et min_s={min_samples}, n_cl={n_classes} et bruit={bruit}')

# print(dbm.labels_)

# plt.scatter(array_1[:,0], array_1[:,1], c='b')
# plt.scatter(array_2[:,0], array_2[:,1], c='r')
plt.scatter(train_data[:,0], train_data[:,1], c=dbm.labels_, cmap=plt.get_cmap('Set1'))
plt.show()
```

```python
dbm = DBSCAN(eps=1.8, min_samples=4)

dbm.fit(train_data)
n_classes = len(list(set(dbm.labels_)))
bruit = len([i for i in dbm.labels_ if i == -1])
print(f'Pour eps={eps} et min_s={min_samples}, n_cl={n_classes} et bruit={bruit}')

# print(dbm.labels_)

# plt.scatter(array_1[:,0], array_1[:,1], c='b')
# plt.scatter(array_2[:,0], array_2[:,1], c='r')
plt.scatter(train_data[:,0], train_data[:,1], c=dbm.labels_, cmap=plt.get_cmap('Set1'))
plt.show()
```
